对于上述类型数据集,可进行如下实现:
```python

class Dataset03(SegmentationSet):
    def __init__(self,data_dir):
        super().__init__(data_dir,description="Dataset 003")

    def _get_all_indentitys(self):
        return listdir(self.data_dir)

    def _id_to_img(self,case_id):
        return os.path.join(case_id,"data.nii.gz")

    def _id_to_seg(self,case_id):
        return os.path.join(case_id,"label.nii.gz")
        
```
